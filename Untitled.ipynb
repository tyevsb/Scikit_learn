# scikit project


```python
import pandas as pd
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
#x_feature matrix
x = heart_disease.drop("target", axis = 1)
#y lable 
y = heart_disease["target"]
```

```python
#choosing the model from scikit-learn
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
#split the data into training set and test set using sklearn model
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf.fit(x_train,y_train)
```

```python
# making a prediction
import numpy as np
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Evaluate the model
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)  #85% of accuracy 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
 #improving the model using the number of estimator
np.random.seed(42)
for i in range(1,100,1):
    print(f"trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"model accuracy on test set: {clf.score(x_test, y_test)*100:.2f}%")
    print(" ")
```

```python
#save the model and load it
import pickle
pickle.dump(clf, open("random forest model.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random forest model.pkl", "rb"))
loaded_model.score(x_test,y_test)
```
